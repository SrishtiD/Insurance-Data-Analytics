```python
##Step 1: Data Engineering 

import pandas as pd

cars_df = pd.read_csv("Cars.csv")
customers_df = pd.read_csv("Customers.csv")
households_df = pd.read_csv("Households.csv")

#Checking for nulls
# print(cars_df.isnull().sum())
# print(customers_df.isnull().sum())
# print(households_df.isnull().sum())

#Count of data
# print(len(cars_df))
# print(len(customers_df))
# print(len(households_df))

#Remove unwanted columns
cars_df = cars_df.loc[:, ~cars_df.columns.str.contains('^Unnamed')]
households_df = households_df.loc[:, ~households_df.columns.str.contains('^Unnamed')]
customers_df = customers_df.loc[:, ~customers_df.columns.str.contains('^Unnamed')]

#Handling missing values
cars_df.dropna(inplace=True)
#customers_df.dropna(inplace=True)
households_df.dropna(inplace=True)

#Remove Null Rows
cars_df.dropna(how='all', inplace=True)
households_df.dropna(how='all', inplace=True)
customers_df.dropna(how='all', inplace=True)

#Remove Duplicates
cars_df.drop_duplicates(inplace=True)
customers_df.drop_duplicates(inplace=True)
households_df.drop_duplicates(inplace=True)

#Comapring count of rows after data cleaning
# print(len(cars_df))
# print(len(customers_df))
# print(len(households_df))

# Rename CAR ID to CAR_ID in the CARS dataset
cars_df.rename(columns={'Car ID': 'CAR_ID'}, inplace=True)

# Remove duplicates in key columns before merging
cars_df = cars_df.drop_duplicates(subset='CAR_ID')
households_df = households_df.drop_duplicates(subset='CAR_ID')
customers_df = customers_df.drop_duplicates(subset='CUST_ID')


# Ensure 'HH Start Date' and 'Date of Birth' columns are in datetime format
# Convert 'HH Start Date'
households_df['HH Start Date'] = pd.to_datetime(households_df['HH Start Date'], format='%m/%d/%y', errors='coerce', exact=False)
households_df['HH Start Date'] = households_df['HH Start Date'].apply(lambda x: x if x.year < pd.Timestamp.now().year else pd.Timestamp(x.year - 100, x.month, x.day))
# Convert 'Date of Birth'
customers_df['Date of Birth'] = pd.to_datetime(customers_df['Date of Birth'], format='%m/%d/%y', errors='coerce', exact=False)
customers_df['Date of Birth'] = customers_df['Date of Birth'].apply(lambda x: x if x.year < pd.Timestamp.now().year else pd.Timestamp(x.year - 100, x.month, x.day))


#Joining the datasets
combined_df = pd.merge(cars_df, households_df, on="CAR_ID", how = 'inner')
combined_df = pd.merge(combined_df, customers_df, on="CUST_ID", how = 'inner')


#Schema of final output along with record count 
print(combined_df.info())
print(f"Record count: {len(combined_df)}")

#Saving Final Results in parquet format
combined_df.to_parquet("combined_data.parquet")

```
